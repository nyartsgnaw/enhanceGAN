```python
import datetime
import os
import sys
try:
	CWDIR = os.path.abspath(os.path.dirname(__file__))
except:
	CWDIR = os.getcwd()
sys.path.append(CWDIR+'/lib/utils')
sys.path.append(CWDIR+'/lib/DCGAN')

from prepare_data import *
from model import *



def start(exp):
	start_time = datetime.datetime.now()
	addr_train='./data/train/train_dataset.csv'
	addr_test='./data/test/df_test.csv'

	X_encode = 'onehot'
	Y_encode = 'twosided'

	pdata = prepare_data(allele=exp['allele'],\
						n_mer =exp['n_mer'] ,\
	                    test_measurement = exp['test_measurement'],\
						iedb_ref = exp['iedb_ref'])
	X_train, Y_train = pdata.build_train(addr_train,X_encode=X_encode,Y_encode=Y_encode)
	X_test, Y_test = pdata.build_test(addr_test,X_encode=X_encode,Y_encode='class',Y_cutoff=exp['ic50_cutoff'])

	GAN = go_GAN(allele = exp['allele'],\
				n_mer = exp['n_mer'],\
				n_epoch=200,\
				X_encode = X_encode,\
				Y_encode = Y_encode,\
				model_id = 'GAN_T'+str(exp['test_id']),\
				G_input_dim= 32,\
				D_output_dim= 3,\
				batch_size = exp['batch_size'],\
				D_class_weight={'true':1.2,'fake':1,'false':1.2},\
				G_class_weight={'true':1,'fake':1,'false':1})

	adam=Adam(lr=exp['lr_G'], beta_1=0.9 ,decay=0.01)
#	sgd = SGD(lr=0.01, decay=0.01, momentum=0.9, nesterov=True)
#	rmsprop1 = RMSprop(lr=0.003, rho=0.9, epsilon=1e-08, decay=0.0)
	rmsprop2 = RMSprop(lr=exp['lr_D'], rho=0.9, epsilon=1e-08, decay=0.01)
	G = GAN.create_G()
	G.compile(loss=exp['loss_G'], optimizer = adam, metrics=['accuracy','mae'])
	D = GAN.create_D()
	D.compile(loss=exp['loss_D'], optimizer = rmsprop2, metrics=['mae'])   #this will lead to poor performance of Discriminator at about epoch 39
	output =GAN.train_GAN(X_train=X_train,Y_train=Y_train,X_test=X_test,Y_test=Y_test,G=G,D=D)
	exp['n_tested']+=1
	exp['start_time'] = start_time
	exp = dict(exp)
	exp.update(output['best_row'])
	return exp
```

```python
if __name__ == '__main__':
	exp_addr = './experiments/exp_logs.csv'
	df = pd.read_csv(exp_addr)
	for i in range(df.shape[0]):
#		if df.iloc[i]['n_tested'] == 0:
			exp = df.iloc[i]
			print(exp)
			exp = start(exp)
			df.loc[i,:] = pd.Series(exp)
			df.to_csv(exp_addr,index=False)
```
